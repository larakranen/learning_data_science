```python
import pandas
```

```python
data = pandas.read_csv("cleaned_animal_disease_prediction.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["Animal_Type"] = encoder.fit_transform(data["Animal_Type"])

```

```python
data
```

```python
data.columns
```

```python
columns_to_encode = ["Breed", "Gender", "Symptom_1", "Symptom_2", "Symptom_3", "Symptom_4"]
for column in columns_to_encode:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])
```

```python
data
```

```python
data = data.drop(labels=["Appetite_Loss", "Vomiting", "Diarrhea", "Coughing", "Labored_Breathing", "Lameness", "Skin_Lesions", "Nasal_Discharge", "Eye_Discharge"], axis=1)
```

```python
def day_converter(duration):
    number, time_unit = duration.split(" ")
    number = int(number)
    if time_unit == "week" or time_unit == "weeks":
        number = number * 7
    elif time_unit == "day" or time_unit == "days":
        number = number
    else:
        print("I found something else:", time_unit)
    return number
```

```python
data["Duration"] = data["Duration"].apply(day_converter)
```

```python
data
```

```python
data
```

```python
def degree_remover(temp):
    number = float(temp.removesuffix("°C"))
    return number
```

```python
data["Body_Temperature"] = data["Body_Temperature"].apply(degree_remover)
```

```python
data
```

```python
save_encoder = LabelEncoder()
data["Disease_Prediction"] = save_encoder.fit_transform(data["Disease_Prediction"])
```

```python
data
```

```python
input = data.drop("Disease_Prediction", axis=1)
```

```python
input
```

```python
output = data["Disease_Prediction"]
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_train, input_test, output_train, output_test = train_test_split(input, output, test_size=0.2, random_state=42)
```

```python
input_train
```

```python
input_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(input_train, output_train)
```

```python
output_predicted = classifier.predict(input_test)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(output_test, output_predicted)
```

```python
print(report)
```

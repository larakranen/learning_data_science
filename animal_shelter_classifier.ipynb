```python
import pandas
```

```python
data = pandas.read_csv("..\project\Austin_Animal_Center_Outcomes.csv")
```

```python
data = data[data["Animal Type"].isin(["Dog", "Cat"])]
```

```python
data = data[~data["Outcome Type"].isin(["Died", "Euthanasia", "Rto-Adopt", "Disposal", "Missing", "Relocate"])]
```

```python
data = data[~data["Outcome Type"].isna()]
```

```python
#ocount = data["Outcome Type"].value_counts()
```

```python
#ocount
```

```python
data = data.drop(labels=["Date of Birth", "Animal ID", "Name", "DateTime", "MonthYear", "Outcome Subtype"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
outcome_encoder = LabelEncoder()
data["Outcome Type"] = outcome_encoder.fit_transform(data["Outcome Type"])
columns_to_encode = ["Animal Type", "Sex upon Outcome", "Breed", "Color"]
for column in columns_to_encode:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])
```

```python
mask = data["Age upon Outcome"].apply(type) == str
data = data[mask]
```

```python
def age_converter(age):
    number, time_unit = age.split(" ")
    if time_unit == "day" or time_unit == "days":
        return int(number)
    elif time_unit == "week" or time_unit == "weeks":
        return int(number) * 7
    elif time_unit == "month" or time_unit == "months":
        return int(number) * 30
    elif time_unit == "year" or time_unit == "years":
        return int(number) * 365
    else:
        print("There is an unknown time unit:", time_unit)
```

```python
data["Age upon Outcome"] = data["Age upon Outcome"].apply(age_converter)
```

```python
input = data.drop("Outcome Type", axis=1)
output = data["Outcome Type"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_train, input_test, output_train, output_test = train_test_split(input, output, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier, ExtraTreesClassifier
```

```python
classifier = GradientBoostingClassifier(max_depth=7, n_estimators=200)
classifier.fit(input_train, output_train)
```

```python
from sklearn.metrics import classification_report
```

```python
output_predicted = classifier.predict(input_test)
report = classification_report(output_test, output_predicted)
print(report)
```
